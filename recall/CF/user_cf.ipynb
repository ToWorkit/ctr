#### 基于用户的协同过滤

#### 数据集格式 {user_id: {item_id: rating}}

```python
# 电影评分集
data_path = "F:\\tmp\\u.data"
```

```python
# 转为DataFrame
udata = pd.read_csv(data_path, sep = "\t", header = None, names = ['user_id', 'item_id', 'rating', 'timestamp'])
udata.head()
```

```python
# 字典类型存储
train = dict()
```

```python
# iterrows返回值（index，row）generator 
for _, row in udata.iterrows():
    # 取出数据
    user_id = str(row['user_id'])
    item_id = str(row['item_id'])
    rating = row['rating']
    # 按照数据格式对user_id初始化（没有默认为-1）
    # 一个用户可能会对多个电影打分
    if train.get(user_id, -1) == -1:
        train[user_id] = dict()
    # 赋值
    train[user_id][item_id] = rating
```

### 获取到相似用户集之后开始构建推荐数据集

#### 1、建立item -> users 的倒排表

```python
# u -> user_id
# i -> item_id
item_users = dict()
for u, items in train.items():
    for i in items.keys():
        if i not in item_users:
            # 去重
            item_users[i] = set()
        item_users[i].add(u)
```

```python
# item_id 242
# item下的用户
print(item_users["242"])
```

#### 2、计算相似user共同的物品数量，优化热门物品的处理(对其进行衰减)

```python
# 用户都打分过的item数量
C = dict()
# 存储单个用户打分过的Item数量
N = dict()
```

```python
for i, users in item_users.items():
    for u in users:
        if N.get(u) == None:
            N[u] = 0
        N[u] += 1
        if C.get(u) == None:
            C[u] = dict()
            
        for v in users:
            if u == v:
                continue
                
            elif C[u].get(v) == None:
                C[u][v] = 0
                
            C[u][v] += 1
            # 对爆款商品进行衰减处理 log 值越大越小(平滑处理)
#             C[u][v] += 1 / math.log(1 + len(users))
```

#### 3、最终的用户相似度矩阵

```python
W = dict()
for u, related_users in C.items():
    if W.get(u, -1) == -1:
        W[u] = dict()
    for v, cuv in related_users.items():
        # jarcard 相似度  交集 / 当前用户和相似用户的物品个数乘积再开方
        W[u][v] = cuv / math.sqrt(N[u] * N[v] * 1.0)
```

```python
import operator
# 按相似度（0是用户，1是相似度）降序排列取前十
# 相似用户
sorted(W['504'].items(), key = operator.itemgetter(1), reverse = True)[:10]
```

#### 4、根据相似用户构建相似物品集合（推荐物品集）

```python
def recommend(user, train, w, k):
    rank = dict()
    # 当前用户物品集(只要item_id)
    interacted_items = train[user].keys()
    
    # 当前用户排序后的相似用户数据集
    sort_items = sorted(w[user].items(), key = operator.itemgetter(1), reverse = True)[0:k]
    
    for v, wuv in sort_items:
        # 相似用户的物品集（item_id, rating）
        for i, rvi in train[v].items():
            # 过滤掉当前用户打分过的物品
            if i in interacted_items:
                continue
            elif rank.get(i) == None:
                rank[i] = 0
            # 相似度 * 打分
            rank[i] += wuv * rvi
    # item_id: 相似度 * 打分
    return rank
```

```python
rec_item_list = recommend('196', train, W, 10)
```

```python
# user_id为196的用户相似物品推荐集合top10
sorted(rec_item_list.items(), key = operator.itemgetter(1), reverse = True)[:10]
```
