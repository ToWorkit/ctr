```python
PATH = "F:\\badou\\tmp\\data\\text\\Cut_Finish_jieba.csv"
```

```python
model = Doc2Vec.load("F:\\badou\\tmp\\data\\text\\doc2vec_stock.model")
x = pd.read_excel(PATH,encoding = 'utf-8')
article = x['內容'].tolist()
```

```python
def cos_sim(vector1, vector2): 
    '''计算余弦相似度'''
    cos=np.dot(vector1,vector2)/(np.linalg.norm(vector1)*(np.linalg.norm(vector2)))
    return cos
```

```python
# 文章向量化
def count_vector(model, words): 
    '''
    依据模型参数(size=500)设定容器
    模型
    文章(分词后的)
    '''
    vect_list = []
    vec = np.zeros(500) 
    for w in words:  
        try:    
            # 从模型当中获取数据
            vec = vec + model.wv[w]
        except:
            continue
    return vec
```

```python
model.wv['鴻海']
```

### 获取每篇文章和给定词语的余弦相似度

```python
for i in article:
    i = i.split(' ')
    vec = count_vector(model,i)
    cos = cos_sim(vec,model.wv['鴻海'])
    print(cos)
```
