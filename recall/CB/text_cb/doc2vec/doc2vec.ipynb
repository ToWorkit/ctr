```python
import jieba
import pandas as pd
from gensim.models.doc2vec import Doc2Vec,TaggedDocument
```

```python
PATH = "F:\\badou\\tmp\\data\\text\\Cut_Finish_jieba.csv"
```

```python
article = pd.read_excel(PATH)
```

```python
sentences = article['內容'].tolist()
```

```python
split_sentences = []

for i in sentences:
    split_sentences.append(i.split(' '))
```

```python
split_sentences
```

```python
documents = [TaggedDocument(doc, [i]) for i, doc in enumerate(split_sentences)]
documents
```

```python
model = Doc2Vec(documents, size=500, window=5, min_count=5, workers=4, epoch=5000)
model.save("F:\\badou\\tmp\\data\\text\\doc2vec_stock.model")
```
